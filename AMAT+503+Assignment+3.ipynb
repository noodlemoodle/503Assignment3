```python
from numpy import *
from matplotlib.pyplot import *
from scipy.signal import convolve2d
from sklearn.preprocessing import normalize

def ff(x,y):
    return (x**2 + y**2 <1)

def gg(x,y):
    return (abs(x)<.5)*(abs(y)<.3)

Delta = .05
x = linspace(-2,2,round(4/Delta))
y = linspace(-2,2,round(4/Delta))

ffimg = zeros((size(x),size(y)))
ggimg = zeros((size(x),size(y)))
for j in range(size(x)):
    for k in range(size(y)):
        ffimg[j,k]=ff(x[j],y[k])
        ggimg[j,k]=gg(x[j],y[k])

result = convolve2d(ffimg,ggimg)
result = normalize(result, axis = 1, norm = 'l2') 

subplot(1,3,1)
extent = min(x), max(x), min(y), max(y)
imshow(ffimg, extent=extent)
subplot(1,3,2)
imshow(ggimg, extent=extent)
subplot(1,3,3)
imshow(result, extent=(-4,4,-4,4))
show()
```

```python
from numpy import *
from matplotlib.pyplot import *
from scipy.signal import convolve2d

def ff(x,y):
    return (x**2 + y**2 <1)

def gg(x,y):
    return (abs(x)<.5)*(abs(y)<.3)

gs = [[1, 1, 1], [1, 1, 1], [1, 1, 1]]
hs = [[0, 1, 0], [1, -4, 1], [0, 1, 0]]

Delta = .05
x = linspace(-2,2,round(4/Delta))
y = linspace(-2,2,round(4/Delta))

ffimg = zeros((size(x),size(y)))
ggimg = zeros((size(x),size(y)))

g = zeros((size(x),size(y)))
h = zeros((size(x),size(y)))

midX = (int)(size(x)/2)
midY = (int)(size(y)/2)

for i in range(3):
    g[midX - 1, midY + i - 1] = 1
    g[midX, midY + i - 1] = 1
    g[midX + 1, midY + i - 1] = 1

h[midX - 1, midY - 1] = 0
h[midX - 1, midY] = 1
h[midX - 1, midY + 1] = 0

h[midX, midY - 1] = 1
h[midX, midY] = -4
h[midX, midY + 1] = 1

h[midX + 1, midY - 1] = 0
h[midX + 1, midY] = 1
h[midX + 1, midY + 1] = 0


for j in range(size(x)):
    for k in range(size(y)):
        ffimg[j,k]=ff(x[j],y[k])
        ggimg[j,k]=gg(x[j],y[k])

# Convolution of circle image with filter g
r1 = convolve2d(ffimg, gs)

# Convolution of circle image with filter h
r2 = convolve2d(ffimg, hs)

# Convolution of rectangle image with filter g
r3 = convolve2d(ggimg, gs)

# Convolution of rectangle image with filter h
r4 = convolve2d(ggimg, hs)

subplot(1,3,1)
extent = min(x), max(x), min(y), max(y)
imshow(ffimg, extent=extent)
subplot(1,3,2)
imshow(gs, extent=extent)
subplot(1,3,3)
imshow(r1, extent=(-4,4,-4,4))
subplots_adjust(hspace = 0.5, wspace = 0.5)
show()

subplot(1,3,1)
extent = min(x), max(x), min(y), max(y)
imshow(ffimg, extent=extent)
subplot(1,3,2)
imshow(hs, extent=extent)
subplot(1,3,3)
imshow(r2, extent=(-4,4,-4,4))
subplots_adjust(hspace = 0.5, wspace = 0.5)
show()

subplot(1,3,1)
extent = min(x), max(x), min(y), max(y)
imshow(ggimg, extent=extent)
subplot(1,3,2)
imshow(gs, extent=extent)
subplot(1,3,3)
imshow(r3, extent=(-4,4,-4,4))
subplots_adjust(hspace = 0.5, wspace = 0.5)
show()

subplot(1,3,1)
extent = min(x), max(x), min(y), max(y)
imshow(ggimg, extent=extent)
subplot(1,3,2)
imshow(hs, extent=extent)
subplot(1,3,3)
imshow(r4, extent=(-4,4,-4,4))
subplots_adjust(hspace = 0.5, wspace = 0.5)
show()
```

## Result

The filters are highlighting the outline of the shapes, and the result of the convolution with h is the negative of the convolution with g.
As the value of delta gets smaller, the outline becomes thinner.

```python
from numpy import *
from matplotlib.pyplot import *


def f1(t):
    return (t>-1)*(t<1)
    
def f2(t):
    return (t > -1) * (t < 1) * t

def f3(t):
    return 1 - (t <= -1) * (t >= 1) - (t > -1) * (t < 1) * t * t

def f4(t, sigma):
    return exp(-(t*t)/(sigma**2))

Delta_t = .01
t = linspace(-2,2,round(4/Delta_t))

def f1hat(w):
    return Delta_t*dot(f1(t),exp(-2*pi*1j*w*t))

def f2hat(w):
    return Delta_t*dot(f2(t),exp(-2*pi*1j*w*t))

def f3hat(w):
    return Delta_t*dot(f3(t),exp(-2*pi*1j*w*t))

def f4hat(w, sigma):
    return Delta_t*dot(f4(t, sigma),exp(-2*pi*1j*w*t))

## Prepare the output samples of the Fourier transform
Delta_w = .01

L = 10
w1 = linspace(-L,L,round(20/Delta_w))
w2 = linspace(-1,1,round(20/Delta_w))
w3 = linspace(-1,1,round(20/Delta_w))
w4 = linspace(-1,1,round(20/Delta_w))

f1hat_out = zeros(size(w),dtype=complex_)
f2hat_out = zeros(size(w2),dtype=complex_)
f3hat_out = zeros(size(w3),dtype=complex_)
f4ahat_out = zeros(size(w4),dtype=complex_)
f4bhat_out = zeros(size(w4),dtype=complex_)
f4chat_out = zeros(size(w4),dtype=complex_)
for k in range(size(w)):
    f1hat_out[k] = f1hat(w1[k])
    f2hat_out[k] = f2hat(w2[k])
    f3hat_out[k] = f3hat(w3[k])
    f4ahat_out[k] = f4hat(w4[k], 0.5)
    f4bhat_out[k] = f4hat(w4[k], 1)
    f4chat_out[k] = f4hat(w4[k], 2)

f, arr = subplots(4, 1)
arr[0].plot(w, real(f1hat_out))
arr[0].set_title('Plot of f1(t) = 1')
arr[1].plot(w, real(f2hat_out))
arr[1].set_title('Plot of f2(t) = t')
arr[2].plot(w, real(f3hat_out))
arr[2].set_title('Plot of f3(t) = 1 - t^2')
arr[3].plot(w, real(f4ahat_out))
arr[3].plot(w, real(f4bhat_out))
arr[3].plot(w, real(f4chat_out))
arr[3].set_title('Plot of f4(t) = exp(-t^2/sigma^2) with sigma = 0.5, 1, and 2')



f.subplots_adjust(hspace = 0.5, wspace = 0.4)
f.set_figwidth(8)
f.set_figheight(14)
show()
```
